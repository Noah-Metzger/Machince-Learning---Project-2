```python
import pandas as pd
import numpy as np
```

```python
glass = pd.read_csv('Data/glass.csv', header=None)
glass = glass.drop([0],axis=1)
```

```python
folds = np.array_split(glass, 10)

train = pd.DataFrame()
test = pd.DataFrame()

for i, fold in enumerate(folds):
    if i == 0:
        test = test.append(fold)
    else:
        train = train.append(fold)
```

```python
def lpNorm(x, y, p):
    sum = 0
    for i in range(len(x)):
        sum += pow(x[i]-y[i], p)
    return pow(sum, (1/p))

# def predict(test, train, test_y, train_y):
print(test.loc[0])
print(test.loc[1])

print(lpNorm(np.array(test.loc[0]), np.array(test.loc[1]), 2))
```

```python
def predict(test, train, train_y, k):
    neighbors = []
    classes = []
    for i, row in train.iterrows():
        dist = lpNorm(np.array(row), np.array(test), 2)
        if len(neighbors) == 0:
            neighbors.append(dist)
            classes.append(train_y[i])
        else:
            print(dist)
            isPut = False
            for j in range(0,len(neighbors)):
                if neighbors[j] > dist:
                    neighbors.insert(j, dist)
                    classes.insert(j, i)
                    isPut = True
                    break
            
        if len(neighbors) > k:
            neighbors = neighbors[0:len(neighbors) - 1]
            classes = classes[0:len(classes) - 1]
        print(classes)
        print(neighbors)

predict(train.loc[22], test, test_y, 5)
```
